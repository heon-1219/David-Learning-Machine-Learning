```python
def preprocessing(df: pd.DataFrame) -> np.array:
  Target = df.Symbol.values[0]
  target_df = df[df['Symbol'] == Target]

  close_data = np.array(target_df.Close.values)
  y_data = np.array(target_df.Close.values)
  len(close_data)
  #study the gradient of every 2 days
  num_sets = int((close_data.shape[0] / 2))

  total_gradient_data = np.zeros(num_sets)

  added = 0
  index = 0

  for i in range(1, len(close_data)) :
    added = close_data[i] - close_data[i-1]
    total_gradient_data[index] = added
    if index < num_sets -1:
      index += 1

  bidaily = np.zeros(num_sets)
  for i in range(0, num_sets):
    bidaily[i] = target_df.Open.values[i+2]

  #total gradient data = x data, y data
  return total_gradient_data, bidaily, y_data
```

```python
tgd, bidaily, y = preprocessing(df)
```
